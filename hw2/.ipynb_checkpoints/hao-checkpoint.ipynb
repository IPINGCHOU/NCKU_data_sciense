```python
#import data
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import random
df=pd.read_csv("data\cdata.csv")
```

```python
#choose central points by kmeans++
x=np.array(df["x"])
y=np.array(df["y"])
num=np.random.choice(range(1,301),1)
centralx=x[num]
centraly=y[num]
old=np.array(df["cluster"])
for i in range(3):
    distance=[]
    distanceid=pd.DataFrame()
    for j in range(i+1): 
        distanceid[j]=((x-centralx[j])**2+(y-centraly[j])**2)**(0.5)
    for i in range(len(x)):
        distance.append(sum(distanceid.iloc[i]))
    newnum=distance.index(max(np.delete(distance,num)))
    num=np.append(num,newnum)
    centralx=np.append(centralx,x[newnum])
    centraly=np.append(centraly,y[newnum])
#kmeans
matr=pd.DataFrame()
cen=pd.DataFrame()
cen[0]=list([centralx,centraly])
k=0
while True:
    dif=pd.DataFrame()
    new=[]
    #distance calculating
    for i in range(4):
        dif[i]=((x-centralx[i])**2+(y-centraly[i])**2)**(0.5)
    #grouping
    for i in range(len(x)):
        new.append(dif.iloc[i].idxmin())
    if(sum(new==old)==len(x)):
        break
    else:
        matr[k]=new
    old=np.array(new)
    k+=1
    #central shifting
    for i in range(4):
        centralx[i]=sum(x[old==i])/len(x[old==i])
        centraly[i]=sum(y[old==i])/len(y[old==i])
    cen[k]=list([centralx,centraly])
    print(cen[k]) 
print(cen)
row=math.ceil(k/3)
plt.rcParams["figure.figsize"]=[15,5*row]
for i in range(k):
    p1=plt.subplot(row,3,i+1)
    p1.scatter(x[matr[i]==0],y[matr[i]==0],color="red",label='A')
    p1.scatter(x[matr[i]==1],y[matr[i]==1],color="blue",label='B')
    p1.scatter(x[matr[i]==2],y[matr[i]==2],color="orange",label='C')
    p1.scatter(x[matr[i]==3],y[matr[i]==3],color="green",label='D')
    p1.scatter(cen[i][0],cen[i][1],marker="x",color="black",label='Central',s=300)
    p1.set_xlabel('x')
    p1.set_ylabel('y')
    p1.legend()
    
```

```python
plt.rcParams['figure.figsize'] = [9,6]
SSE=[]
for i in range(2,51,1):
    #choose central points by kmeans++
    x=np.array(df["x"])
    y=np.array(df["y"])
    np.random.seed(0)
    num=np.random.choice(range(1,301),1)
    centralx=x[num]
    centraly=y[num]
    for f in range(i-1):
        distance=[]
        distanceid=pd.DataFrame()
        for j in range(f+1): 
            distanceid[j]=((x-centralx[j])**2+(y-centraly[j])**2)**(0.5)
        for k in range(len(x)):
            distance.append(sum(distanceid.iloc[k]))
        newnum=distance.index(max(np.delete(distance,num)))
        num=np.append(num,newnum)
        centralx=np.append(centralx,x[newnum])
        centraly=np.append(centraly,y[newnum])
    old=np.array(df["cluster"])
    #kmeans
    while True:
        dif=pd.DataFrame()
        new=[]
        mindif=[]
        #distance calculating
        for j in range(i):
            dif[j]=((x-centralx[j])**2+(y-centraly[j])**2)**(0.5)
        #grouping
        for t in range(len(dif[0])):
            new.append(dif.iloc[t].idxmin())
            mindif.append(dif.iloc[t].min())
        #SSE calculating
        if(sum(new==old)==len(dif[0])):
            SSE.append(np.sum(np.array(mindif)**(2)))
            break
        old=np.array(new)
        #central shifting
        for j in range(i):
            centralx[j]=np.sum(x*[old==j])/sum(old==j)
            centraly[j]=np.sum(y*[old==j])/sum(old==j)
#plot
plt.plot(range(2,51,1), SSE)
plt.xlabel('Number of k')
plt.ylabel('Sum of square error')
plt.show()
```
