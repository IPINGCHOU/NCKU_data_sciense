```python
# packages imported
import pandas as pd
import numpy as np
import mord
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn import linear_model, preprocessing

# import data
# file_name 檔案路徑
file_name = "original_data/training_data.csv"
data = pd.read_csv(file_name, header = 0)
train = pd.DataFrame(data)
file_name_test = 'original_data/test_data.csv'
data = pd.read_csv(file_name_test, header = 0)
test = pd.DataFrame(data)
```

```python
total_text = pd.concat([train['text'], test['text']], ignore_index = True)
len(total_text)
# train 0~7996
# test  7997~9999
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_vectorizer.fit(total_text)
vectorizer = CountVectorizer()
vectorizer.fit(total_text)

total_text_freq = vectorizer.fit_transform(total_text)

transformer = TfidfTransformer(sublinear_tf=True, use_idf=True)
total_text_tfidf = transformer.fit_transform(total_text_freq)


# tfidf 矩陣
# print(total_text_tfidf.toarray())
# 總出現次數矩陣
# print(train_freq.toarray())
# features
# print(train_word)
```

```python
total_text_tfidf.shape
```

```python
# TruncatedSVD
n_components = 3000
n_iter = 5
from sklearn.decomposition import TruncatedSVD
TSVD = TruncatedSVD(n_components = n_components, n_iter = n_iter)
total_text_after_TSVD = TSVD.fit_transform(total_text_tfidf)
```

```python
total_text_tfidf.shape
```

```python
# 拆回 train test
train_text_after_TSVD = total_text_after_TSVD[0:7997,:]
test_text_after_TSVD = total_text_after_TSVD[7997:10000,:]

# np.savetxt("train_text_after_TSVD.csv", train_text_after_TSVD, delimiter=",")
# np.savetxt("test_text_after_TSVD.csv", test_text_after_TSVD, delimiter=",")

print(train_text_after_TSVD.shape)
print(test_text_after_TSVD.shape)
```

```python
X = train_text_after_TSVD
y = train['stars']
```

```python
# Ordinal logistic with Lasso
clf = mord.LogisticAT(alpha = 1.)
clf.fit(X,y)

#預測
test_pred = clf.predict(test_text_after_TSVD)
review_id = test['review_id']
out = pd.DataFrame()
out[0] = review_id
out[1] = test_pred

out.to_csv('out.csv', header = False, index = False)
```

```python
# Ordinal Logistic with Ridge
ridge = mord.OrdinalRidge()
ridge.fit(X, y)

test_pred = ridge.predict(test_text_after_TSVD)
review_id = test['review_id']
out = pd.DataFrame()
out[0] = review_id
out[1] = test_pred

out.to_csv('out.csv', header = False, index = False)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'C': [0.01, 0.1, 1, 2, 4, 10],
}
# Create a based model
LAD = mord.LAD()
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = LAD, param_grid = param_grid, 
                          cv = 5, verbose = 2)

grid_search.fit(X, y)
```

```python
grid_search.grid_scores_,grid_search.best_params_, grid_search.best_score_
```

```python
# mord LAD
LAD = mord.LAD(C = 1)
LAD.fit(X, y)

test_pred = LAD.predict(test_text_after_TSVD)
review_id = test['review_id']
out = pd.DataFrame()
out[0] = review_id
out[1] = test_pred

out.to_csv('out.csv', header = False, index = False)
```

```python
# RandomForest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf = RandomForestClassifier()
rf.fit(X, y)

predicted = rf.predict(X)
accuracy = accuracy_score(y, predicted)
print(accuracy)

#model predict
test_pred=rf.predict(test_text_after_TSVD)
review_id = test['review_id']
out = pd.DataFrame()
out[0] = review_id
out[1] = test_pred
out.to_csv('out.csv', header = False, index = False)
```

```python
# SVM
from sklearn import svm
clf = svm.SVC(decision_function_shape='ovo')
clf.fit(X, y) 

test_pred=clf.predict(test_text_after_TSVD)
review_id = test['review_id']
out = pd.DataFrame()
out[0] = review_id
out[1] = test_pred
out.to_csv('out.csv', header = False, index = False)
```

```python
# SVR SearchGridCV
from sklearn import svm
from sklearn.model_selection import GridSearchCV
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
    'gamma': [0, 1, 2, 3, 4, 5, 6, 7],
}
# Create a based model
svr = svm.SVR()
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = svr, param_grid = param_grid, 
                          cv = 10, verbose = 2)

grid_search.fit(X, y)
```

```python
grid_search.grid_scores_,grid_search.best_params_, grid_search.best_score_
```

```python
# SVR
from sklearn import svm
svr = svm.SVR(C = 10, gamma = 2)
svr.fit(X, y)

test_pred=svr.predict(test_text_after_TSVD)

for i in range(len(test_pred)):
    if test_pred[i] >= 6:
        test_pred[i] = 5

review_id = test['review_id']
out = pd.DataFrame()
out[0] = review_id
out[1] = test_pred
out.to_csv('out.csv', header = False, index = False)
```

```python
# GLM
from sklearn.linear_model  import LogisticRegression
lr = LogisticRegression()
lr.fit(X, y)

test_pred = lr.predict(test_text_after_TSVD)
review_id = test['review_id']
out = pd.DataFrame()
out[0] = review_id
out[1] = test_pred
out.to_csv('out.csv', header = False, index = False)
```
