```python
# import  package
import numpy as np
import pandas as pd
import tensorflow as tf
import datetime as dt
```

```python
# import data and data after trauncated svd
file_name = "original_data/training_data.csv"
data = pd.read_csv(file_name, header = 0)
train = pd.DataFrame(data)
file_name_test = 'original_data/test_data.csv'
data = pd.read_csv(file_name_test, header = 0)
test = pd.DataFrame(data)

file_name = 'original_data/train_text_after_TSVD.csv'
data = pd.read_csv(file_name, header = None)
TSVD_train = pd.DataFrame(data)
file_name = 'original_data/test_text_after_TSVD.csv'
data = pd.read_csv(file_name, header = None)
TSVD_test = pd.DataFrame(data)


''' Convert to one-hot encoding '''
#將 Y (星數) 變為 dummy variable
y = train['stars']

for i in range(len(y)):
    if y[i] == 1:
        out = np.asarray([0,0,0,0])
    elif y[i] == 2:
        out = np.asarray([1,0,0,0])
    elif y[i] == 3:
        out = np.asarray([1,1,0,0])
    elif y[i] == 4:
        out = np.asarray([1,1,1,0])
    elif y[i] == 5:
        out = np.asarray([1,1,1,1])
    
    if i == 0:
        enc_y = out
    else:
        enc_y = np.row_stack((enc_y, out))
        
enc_y = pd.DataFrame(enc_y)
```

```python
#import packages
from keras.models import Sequential, Model, load_model
from keras.layers.core import Dense, Activation, Dropout
from keras.optimizers import Adam, RMSprop, Adagrad , adadelta, Nadam
from keras.regularizers import l1,l2
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras.layers.core import Dropout
from keras.callbacks import Callback
from keras.layers.advanced_activations import LeakyReLU
from keras.layers import Input, Concatenate
from keras.models import model_from_json

#Self-defined Callbacks
class LossHistory(Callback):
    def on_train_begin(self, logs={}):
        self.loss = []
        self.acc = []
        self.val_loss = []
        self.val_acc = []
    def on_epoch_end(self, batch, logs={}):
        self.loss.append(logs.get('loss'))
        self.val_loss.append(logs.get('val_loss'))
        self.acc.append(logs.get('acc'))
        self.val_acc.append(logs.get('val_acc'))
        
loss_history = LossHistory()

#early stopping
#以 loss為基準  若是loss連續3次無法下降 則停止學習
earlyStopping=EarlyStopping(monitor= 'loss', patience= 3)
```

```python
#DNN model
#left model
input_1 = Input(shape = (2000,))
sbs = Dense(128, )(input_1)
sbs = LeakyReLU(0.02)(sbs)
sbs = Dense(256, )(sbs)
sbs = LeakyReLU(0.02)(sbs)
sbs = Dense(512, )(sbs)
sbs = LeakyReLU(0.02)(sbs)
sbs = Dense(1024, )(sbs)
sbs = LeakyReLU(0.02)(sbs)
sbs = Dense(2048, )(sbs)
sbs = LeakyReLU(0.02)(sbs)
sbs = Dense(1024, )(sbs)
sbs = LeakyReLU(0.02)(sbs)
sbs = Dense(512, )(sbs)
sbs = LeakyReLU(0.02)(sbs)
sbs = Dense(256, )(sbs)
sbs = LeakyReLU(0.02)(sbs)
sbs = Dense(128, )(sbs)
sbs = LeakyReLU(0.02)(sbs)


predictions = Dense(4, activation = 'sigmoid')(sbs)

model_fin = Model(inputs = [input_1], outputs = predictions)

adam = Adam(lr=0.0001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model_fin.compile(loss = 'mean_squared_error',
                    optimizer = adam,
                    metrics=['accuracy'])
```

```python
history_adam = model_fin.fit([TSVD_train], enc_y,
                           batch_size = 2048,
                           epochs = 500,
                           verbose = 1,
                           shuffle=True,
                           validation_split=0.2,
                           callbacks=[loss_history,earlyStopping])
```

```python
pred = model_fin.predict([TSVD_test], verbose=1)
```
